3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def transform_experience(raw_data):
    """Метод парсинга опыта работы"""
    if pd.isnull(raw_data) or raw_data == 'Не указано': # если данных нет, оставляем как есть
        return np.nan
    else:
        splitted_data = raw_data[len('Опыт работы '):].split()
        if splitted_data[1] in ['год', 'года', 'лет']:
            months = int(splitted_data[0])*12
            # дальше смотрим есть ли информация о месяцах после годов
            if splitted_data[2].isdigit():
                return months + int(splitted_data[2])
            else:
                return months
        else: # сразу идут месяцы - возвращаем их
            return int(splitted_data[0])

df['Опыт работы (месяц)'] = df['Опыт работы'].apply(transform_experience)
df.drop(['Опыт работы'], axis=1, inplace=True)
print(f'Медианный опыт работы: {df['Опыт работы (месяц)'].median()}')
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']

def parse_city(city_line):
    """Метод парсинга города из строчки «Город, переезд, командировки»"""
    city = city_line.split(',')[0].strip()
    if city in ['Москва', 'Санкт-Петербург']:
        return city
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

def parse_relocation(city_line):
    """Метод распознавания готовности к переезду"""
    if 'не готова к переезду' in city_line or 'не готов к переезду' in city_line:
        return False
    else:
        return True

def parse_trips(city_line):
    """Метод распознавания готовности к командировкам"""
    sections = city_line.split(',') # делим на секции по запятой
    readiness = [sec for sec in sections if 'командировка' in sec] # находим секцию с командировками
    if len(readiness) == 0: # нет такого блока, значит не указано
        return False
    else:
        return not('не' in readiness[0] or 'Не' in readiness[0]) # смотрим по наличию отрицания


df['Город'] = df['Город, переезд, командировки'].apply(parse_city).astype('category')
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(parse_relocation)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(parse_trips)
df.drop(['Город, переезд, командировки'], axis=1, inplace=True)

in_spb = round(df['Город'].value_counts(normalize=True)['Санкт-Петербург']*100)
print(f'В Санкт-Петербурге {in_spb}% соискателей')
ready_relocate_and_trip = df[df['Готовность к переезду'] & df['Готовность к командировкам']].shape[0]
print(f'ОДновременно готовы к переездам и командировкам {round((ready_relocate_and_trip/df.shape[0])*100)}% соискателей')
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Признак занятость
empl_values = sum([x.split(',') for x in df['Занятость'].unique()], [])
empl_unique_values = list(set([x.strip() for x in empl_values])) # получили массив значений ['полный день', 'вахтовый метод'..]
for empl_type in empl_unique_values:
    df[empl_type] = df['Занятость'].str.contains(empl_type)

# Признак график
schedule_values = sum([x.split(',') for x in df['График'].unique()], [])
schedule_unique_values = list(set([x.strip() for x in schedule_values])) # получили массив значений ['полный день', 'вахтовый метод'..]
for schedule_type in schedule_unique_values:
    df[schedule_type] = df['График'].str.contains(schedule_type)

df.drop(['График', 'Занятость'], axis=1, inplace=True)

project_and_volunteer = df[df['проектная работа'] & df['волонтерство']].shape[0]
shift_and_flex = df[df['вахтовый метод'] & df['гибкий график']].shape[0]

print(f'Ищут проектную работу и волонтёрство: {project_and_volunteer}')
print(f'Хотят работать вахтовым методом и с гибким графиком: {shift_and_flex}')
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
import warnings 
warnings.filterwarnings("ignore")

# составим датасет с курсами валют
currencies_df = pd.read_csv('data/ExchangeRates.csv', usecols=['currency', 'date', 'close', 'proportion'])
currencies_df['date'] = pd.to_datetime(currencies_df['date'])
currency_mapping = {
    'грн.': 'UAH',
    'бел.руб.': 'BYN',
    'сум': 'UZS'
    }

# подготавливаем данные в основном датасете
def extranct_currency(salary_line):
    """Метод для получения кода валюты"""
    currency = salary_line.split(' ')[1].strip()
    return currency_mapping.get(currency, currency) # если нет в маппинге, то вернём как было

df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме']).dt.date.astype('datetime64[ns]')
df['ЗП сумма'] = df['ЗП'].apply(lambda x: x.split(' ')[0]).astype('int32')
df['ЗП валюта'] = df['ЗП'].apply(extranct_currency)

# мёржим данные о курсах к основному датасету
df = df.merge(currencies_df, left_on=['ЗП валюта', 'Обновление резюме'], right_on=['currency', 'date'], how='left')
df.fillna({'close':1, 'proportion':1}, inplace=True) # ставим 1 рублю

# проставляем колонку зарплаты в рублях
df['ЗП (руб)'] = df['ЗП сумма'] * df['close'] / df['proportion']
# удаляем промежуточные столбцы
df.drop(['ЗП', 'ЗП сумма', 'ЗП валюта', 'currency', 'date', 'close', 'proportion'], axis=1, inplace=True)

print(f'Желаемая медианная зарплата: {round(df['ЗП (руб)'].median())}')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8)) # два графика рядом
sns.histplot(
    data=df,
    x='Возраст',
    bins=30,
    kde=True, # сглаживание линией
    ax=axes[0]
)
axes[0].set(title='Гистограмма распределения возраста', ylabel='Количество')

sns.boxplot(
    data=df,
    x='Возраст',
    ax=axes[1]
)
axes[1].set(title='Коробчатая диаграмма распределения возраста')
print(f'Мода: {df['Возраст'].mode()[0]}')

def majority_interval(data):
    """Вычисляем интервал величины Q1 - 1,5*iqr; Q3 + 1,5*iqr"""
    iqr = data.quantile(0.75) - data.quantile(0.25) # межквантильный размах
    return np.maximum(data.quantile(0.25) - 1.5*iqr, data.min()), np.minimum(data.quantile(0.75) + 1.5*iqr, data.max())

min, max = majority_interval(df['Возраст'])
print(f'Большинство возрастов лежит в интервале {min}-{max}')
```
